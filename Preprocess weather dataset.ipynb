```python
import pandas as pd
import datetime
import csv
```

##### first choose the wanted study period of the weather data 

```python
#format datetime
df_pickle_weather['DATE']=pd.to_datetime(df_pickle_weather['DATE'], format="%Y-%m-%d")
```

```python
df_pickle_weather_18 = df_pickle_weather[(df_pickle_weather["DATE"] > "2018-7-1") & 
                                         (df_pickle_weather["DATE"] < "2018-8-30")]
#greater than the start date and smaller than the end date
mask_time_18 = (df_pickle_weather['DATE'] >= "2018-7-1") & (df_pickle_weather['DATE'] <= "2018-8-30")
df_pickle_weather_18 = df_pickle_weather.loc[mask_time_18]
```

```python
df_pickle_weather_19 = df_pickle_weather[(df_pickle_weather["DATE"] > "2019-7-1") & 
                                         (df_pickle_weather["DATE"] < "2019-8-30")]
#greater than the start date and smaller than the end date
mask_time_19 = (df_pickle_weather['DATE'] >= "2019-7-1") & (df_pickle_weather['DATE'] <= "2019-8-30")
df_pickle_weather_19 = df_pickle_weather.loc[mask_time_19]
```

```python
#Combine these two dataset together
df_weather = df_pickle_weather_18.append(df_pickle_weather_19)
```

```python
#Fill all NaN value with 0
df_weather.fillna(0)
```

##### Select the wanted attribute for analysis

```python
df_weather = df_weather[["DATE", "AWND", "PRCP", "SNOW", "SNWD", "TMAX", "TMIN"]]
```
